## Phase 3: Streamlit Dashboard Build
Build an interactive dashboard that uses the trained model to predict disruption risk, severity, and give a simple operational recommendation

## Task 1: Set Up Streamlit App
- create main app file (app.py)
- organize project structure:
- model files
- helper functions (if needed)
- make sure required packages are installed

## Task 2: Build Input Panel
Create user inputs based on selected features
Use:
- dropdowns for categorical features
- sliders/number inputs for numeric features
Example inputs:
- Base
- Division
- Total Blocked Hours
- Sequence Duty Days
- Total Flight Time
- Total Legs
- Max Legs Per Day
- Layover
- Assignment Count
Goal:
- keep inputs simple and realistic for an operations user

## Task 3: Load Trained Model(s)
- load saved disruption model
- load severity model (if using one)
- ensure feature order matches training data exactly

## Task 4: Build Prediction Logic
When user clicks Predict:
- collect inputs into a dataframe
- apply same preprocessing used during training
- generate:
    - disruption prediction
    - probability score
    - severity prediction (if disrupted)

## Task 5: Display Results
Disruption Likelihood
Show:
- predicted label: Stable or Disrupted
- probability score
- risk level:
    - Low Risk
    - Medium Risk
    - High Risk

## Task 6: Severity
If disrupted:
- display:
    - Minor
    - Moderate
    - Severe
If not disrupted:
 - show: “No significant disruption expected”

## Task 7: Operational Recommendation
Map predictions to a clear action:
- Low Risk
    - No immediate action needed
- Medium Risk
    - Monitor sequence and downstream connections
- High Risk + Moderate
    - Review connections and prepare intervention
- High Risk + Severe
    - Flag for proactive operational review

## Task 8: Add Simple Visual Design
- clear title + section headers
- use metric-style outputs (clean numbers + labels)
- color cues:
    - green = low risk
    - yellow/orange = medium
    - red = high

## Task 9: Test the App
Test at least:
- one low-risk case
- one medium-risk case
- one high-risk case

Check:
- predictions display correctly
- severity only appears when needed
- recommendations match outputs

## Task 10: Finalize Project Files

Prepare:
- app.py
- saved model files (.pkl / .joblib)
- requirements.txt
- short README:
    - what the project does
    - how to run it
    - what the outputs mean

## Deliverables
- working Streamlit dashboard
- connected trained model(s)
- clear disruption + severity + recommendation outputs
- tested app
- ready for demo or submission
