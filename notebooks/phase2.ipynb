## Phase 2: Modeling & Dashboard Logic
Train the model(s), evaluate performance, and define how predictions will appear in the StreamLit dashboard

## Task 1: Build Modeling Dataset
- split data into:
   - training set
   - test set
- define:
   - X = selected features
   - Y = significant_change for disruption likelihood model

## Task 2: Train Disruption Model
- Start with a simple baseline model, then compare with a stronger tree-based model
    Reccomoded Order:
        - Logistic Regression
        - Random Forest
        - XGBoost or Gradient Boosting if needed
- Because the data is imbalanced:
   - use class_weight = "balanced" when possible
   - stratify train/test split
   - do not judge performance mainly by accuracy

## Task 3: Build Severity Model
- For rows where significant_change = 1:
   - create a second modeling dataset
   - predict arrival_change_class

## Task 4: Evaluate Severity Model
- Look at:
    - class distribution
    - classification report
    - confusion matrix

## Task 5: Finalize Dashboard Prediction Logic
- Decide how model outputs will appear in the dashboard.
- Disruption likelihood output
   - Show:
        - predicted class: stable or disrupted
        - probability score
        - risk level:
            - Low Risk
            - Medium Risk
            - High Risk
        - Suggested logic:
            - Low Risk = probability < 0.20
            - Medium Risk = 0.20 to 0.50
            - High Risk = > 0.50
    - Severity output
        - if disrupted:
            - show predicted severity:
                - minor
                - moderate
                - severe
        - if stable:
            - severity is not applicable

## Task 6: Operational Reccomendation
- Map the output to a simple recommendation:
    - Stable/Low Risk
      - No immediate operational action needed
    - Minor/Medium Risk
      - Monitor sequence and review downstream connections
    - Moderate/High Risk
      - Review connection impact and prep for posible intervention
    - Severe/High Risk
      - Flag for proactive operational review

- trained disruption model

- finalized dashboard prediction logic
